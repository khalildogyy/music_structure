```python
import os
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

mpl.rcParams['font.sans-serif'] = ['Microsoft YaHei']
mpl.rcParams['axes.unicode_minus'] = False

reports_files = os.listdir('../thesis/已完成调研表/')
reports_files
```

```python
data = []
for file in reports_files:
    temp = pd.read_excel('../thesis/已完成调研表/{}'.format(file))
    temp = temp.iloc[2:12, :5]
    temp.columns = ['音乐', '场景1', '场景2', '评价', '情感打分']
    data.append(temp)
data = pd.concat(data)
data
```

```python
# vlog下音乐的情感分布

ss = []
for i in range(10):
    ss.append(data[data['情感打分']==i].shape[0])
    print("score {}, num: {}".format(i, data[data['情感打分']==i].shape[0]))
    

fig, ax = plt.subplots(figsize=(4, 3))
# ax.bar(range(10), ss)
sns.distplot(list(data['情感打分'].values), bins=10, ax=ax)

ax.set_xticks(range(10))
ax.set_xticklabels(range(10), fontsize=12)
ax.set_xlabel(u'情 感 打 分', fontsize=12)
ax.set_yticks([0, 0.05, 0.1, 0.15, 0.2, 0.25])
ax.set_yticklabels([0, 0.05, 0.1, 0.15, 0.2, 0.25], fontsize=12)
ax.set_ylabel(u'占 比', fontsize=12)

plt.tight_layout()
# plt.savefig('/Users/JJjie/Downloads/情感打分统计.pdf')
plt.show()
```

```python
# 不同的vlog场景，音乐情感分布
scens = ['大海 - 沙滩','森林 - 光束','雪地 - 结冰','沙漠 - 戈壁','草原 - 牛羊','天空 - 彩虹']

plotdata = []

sta = np.zeros([6, 10], dtype=int)

for i in range(1, 7):
    temp1 = data[data['场景1']==i]
    temp2 = data[data['场景1']==i]
    tempd = list(temp1['情感打分'].values) + list(temp2['情感打分'].values) 
    temp = [[a, b] for a, b in zip(tempd, [scens[i-1]]*len(tempd))]
    for a, b in temp:
        sta[i-1, int(a)] += 1
    plotdata.append(temp)

stastics = []
for i in range(6):
    for j in range(10):
        stastics.append([i, j, sta[i, j]])
print(stastics)

plotdata = pd.DataFrame(np.concatenate(plotdata), columns=['score', 'scene'])
plotdata['score'] = plotdata['score'].astype(float)
# plotdata

sns.set_style('whitegrid', {'font.sans-serif':['Microsoft YaHei']})
colors = sns.color_palette("RdBu", n_colors=9)
fig, ax = plt.subplots(figsize=(15, 5))
sns.boxplot(x="scene", y="score", data=plotdata, palette=colors, width=0.25, ax=ax)
sns.despine(left=True)

ax.set_ylabel(u'情 感 分 数', fontsize=16)
ax.set_yticklabels(np.arange(0, 10), fontsize=14)
ax.set_xlabel(u'场 景', fontsize=16)
ax.set_xticklabels(scens, fontsize=14, rotation=0)

plt.tight_layout()
plt.savefig('/Users/JJjie/Downloads/emo-sce.pdf', bbox_inches='tight', pad_inches=0.)
plt.show()
```

```python
# 统计词频

import collections
from wordcloud import WordCloud

def getwords(tempdata):
    def split(w):
        s = ','
        if ',' in w:
            s = ','
        elif '，' in w:
            s = '，'
        elif '、' in w:
            s = '、'
        return w.split(s)

    w0 = []
    for w in tempdata:
        w0 += split(w)
    wordset = []
    for w in w0:
        if len(w) == 4:
            wordset += [w[:2], w[2:]]
        else:
            wordset += split(w)
            
    # remove
    for i in range(len(wordset)-1, -1, -1):
        if wordset[i] in ''.split('，'):
            wordset.pop(i)
    return wordset
    
words = getwords(data['评价'])

# 词频统计
word_counts = collections.Counter(words) # 对分词做词频统计
print (word_counts.most_common(10)) # 输出检查

# 词频展示
wc = WordCloud(collocations=True, 
               background_color='white',
               font_path=r'/Library/Fonts/Microsoft/SimHei.ttf', 
               width=1400, 
               height=960, 
               margin=2).generate(" ".join(words))
plt.imshow(wc) # 显示词云
plt.axis('off') # 关闭坐标轴
plt.show() # 显示图像
wc.to_file(r'/Users/JJjie/Downloads/word_all.pdf')
```
