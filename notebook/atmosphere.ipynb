```python
import pandas as pd
import numpy as np

data = pd.read_csv('/Users/JJjie/Downloads/image_data_v2.csv', index_col=0, encoding='gbk')
data
```

```python
months = {'Jan':1, 'Feb':2, 'Mar':3, 'Apr':4, 'May':5, 'Jun':6, 'Jul':7, 'Aug':8, 'Sep':9, 'Oct':10, 'Nov':11, 'Dec':12}
seasons = {'SON':9, 'DJF':12, 'MAM':3, 'JJA':6}

times = []
for i in range(data.shape[0]):
    year = data['year'].iloc[i]
    if year is not np.nan:
        if len(year) > 4:
            times.append('')
            continue
    else:
        times.append('')
        continue
    
    timestr = data['time_tip'].iloc[i]
    if timestr is not np.nan:
        if timestr[0].isdigit():
            t = timestr[:8]
            if ' ' in t:
                t = year + ("%02d01" % seasons[timestr.split(' ')[1]])
        else:
            try:
                t = year + ("%02d01" % months[timestr])
            except:
                t = ''
        times.append(t)
    else:
        times.append(t)
```

```python
data['normaltime'] = pd.Series(times, index=data.index)
data
```

```python
data.to_csv('/Users/JJjie/Downloads/image_data_new.csv', encoding='utf-8')
```
